```ruby
require 'charty'

Charty::Backends.use(:pyplot)
Charty::Backends::Pyplot.activate_iruby_integration
```

```ruby
require 'daru'
df = Daru::DataFrame.new({'a':[1,2,3,4], 'b':[4,5,6,7], 'c':[8, 9, 10, 11]})
```

```ruby
Charty.bar_plot(:a, :b, data: df).render
```

```ruby
Charty.bar_plot(data: df).render
```

```ruby
df = Daru::DataFrame.new({
  'a': Array.new(100) { rand(10..100) },
  'b': Array.new(100) { rand(20..120) },
  'c': Array.new(100) { rand(30..60) }
})

Charty.box_plot(data: df).render
```

```ruby
bubble = charty.to_bubble(:a, :b, :id)
bubble.render('sample')
```

```ruby
curve = charty.to_curve(:a, :b)
curve.render('sample')
```

```ruby
scatter = charty.to_scatter(:a, :b)
scatter.render('sample')
```

```ruby
error_bar = charty.to_error_bar(:a, :b)
error_bar.render('sample')
```

```ruby
hst= charty.to_hst(:a, :b)
hst.render('sample')
```
